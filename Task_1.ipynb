```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a list of random values
values = np.random.randint(0, 101, 20)
# Specified list of values ​​for testing
# values = [88, 26, 71, 56, 34, 93, 4, 53, 87, 97, 23, 18, 54, 48, 78, 40, 70, 89, 4, 59]
# values = [86, 25, 63, 51, 33, 93, 79, 62, 98, 91, 10, 28, 76, 30, 97, 97, 1, 3, 25, 40]
print("Generated list of values:", values)
```

```python
# Function to find the deepest lake
def find_deepest_lake(heights):
    max_depth = 0
    start_index = end_index = -1

    for i in range(1, len(heights) - 1):
        if heights[i] < heights[i - 1] and heights[i] < heights[i + 1]:
            # Determining the boundaries of the lake
            left_index = i
            while left_index > 0 and heights[left_index] <= heights[left_index - 1]:
                left_index -= 1

            right_index = i
            while right_index < len(heights) - 1 and heights[right_index] <= heights[right_index + 1]:
                right_index += 1

            # Calculating the depth of a lake
            left_peak = heights[left_index]
            right_peak = heights[right_index]
            lake_depth = min(left_peak, right_peak) - heights[i]

            if lake_depth > max_depth:
                max_depth = lake_depth
                start_index, end_index = left_index, right_index

    return max_depth, start_index, end_index
```

```python
# Search for the deepest lake
deepest_lake, start_of_lake, end_of_lake = find_deepest_lake(values)
print(f"Depth of the deepest lake: {deepest_lake}")
print(f"Границы озера: from {start_of_lake} to {end_of_lake}")
```

```python
lake_borders = values[start_of_lake:end_of_lake + 1] if start_of_lake != -1 and end_of_lake != -1 else []
print(f"Borders of the lake are: {lake_borders}")
```

```python
# Created graphics
plt.figure(figsize=(10, 5))
lines = plt.plot(values, linestyle='-', color='b')  # Save the line object

# Finding the boundaries of a lake
lower_lake_boundary = min(lake_borders)
upper_lake_boundary = lake_borders[0] if lake_borders[0] < lake_borders[-1] else lake_borders[-1]

# Adding dotted lines of least value
plt.axhline(lower_lake_boundary, color='g', linestyle='--')

# Adding a dotted line and a point for the depth of the largest lake
plt.axhline(upper_lake_boundary, color='g', linestyle='--')

# Create a new line to display the "lake" in red
plt.plot(
  range(start_of_lake, end_of_lake + 1),
  values[start_of_lake:end_of_lake + 1],
  linestyle='-', color='r'
  )

# Create Labels to axes
plt.xlabel('Index')
plt.ylabel('The height of mountains and the depth of  lakesа')

# Add a title
plt.title('Graph of random integer values')

# Show graph
plt.show()
```
